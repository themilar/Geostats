```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator) 
from scipy import stats
import math, random
```

```python
mean_poro,stdv_poro=0.15,0.02
mu_volume, sigma_volume = 13,0.5
min_oilsat,max_oilsat=0.2,0.4
L = 1000
min_poro,max_poro=0.1,0.2
min_volume,max_volume=0,2
```

```python
porosity_value=np.random.normal(mean_poro,stdv_poro,size=L)
volume_value=np.random.lognormal(mu_volume,sigma_volume,size=L)/1.0e6
oilsaturation_value=np.random.uniform(min_oilsat,max_oilsat,size=L)
```

```python
def add_grid():
    plt.gca().grid(True, which='major',linewidth = 1.0); plt.gca().grid(True, which='minor',linewidth = 0.2) # add y grids
    plt.gca().tick_params(which='major',length=7); plt.gca().tick_params(which='minor', length=4)
    plt.gca().xaxis.set_minor_locator(AutoMinorLocator()); plt.gca().yaxis.set_minor_locator(AutoMinorLocator()) # turn on minor ticks
def hist_easyplot(values,value_range,bin_no,xlabel,ylabel,title,color="blue",edgecolor="black",cumulative=False,log=False,weights=None,):
    # function definition for easier plotting of histograms
    plt.hist(values,bins=bin_no,color=color,edgecolor=edgecolor,range=value_range,log=log,weights=weights,cumulative=cumulative)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    add_grid()
```

```python
plt.subplot(131)
hist_easyplot(porosity_value,value_range=[min_poro,max_poro],bin_no=50,log=False,cumulative=False,weights=None,xlabel="Average Porosity (Fraction)",ylabel="",title="Average Porosity Realisations")
plt.ylim(0,60)
plt.subplot(132)
hist_easyplot(volume_value,value_range=[min_volume,max_volume],bin_no=50,log=False,cumulative=False,weights=None,xlabel="Volume ($10^6$ $m^3$)",ylabel="",title="Average Volume Realisations")
plt.ylim(0,120)
plt.subplot(133)
hist_easyplot(oilsaturation_value,value_range=[min_oilsat,max_oilsat],bin_no=50,log=False,cumulative=False,weights=None,xlabel="Oil Saturation Fraction",ylabel="",title="Average Oil Saturation Realisations")
plt.ylim(0,40)
plt.subplots_adjust(left=0.0, bottom=0.0, right=2.0, top=1.2, wspace=0.2, hspace=0.2)
plt.savefig("chart_iterations/montecarlo_realisations4.png",bbox_inches="tight")
plt.show()
```

```python
OIP=porosity_value*volume_value*oilsaturation_value*6.29
hist_easyplot(OIP,value_range=[0,.8],bin_no=50,log=False,cumulative=False,color="green",weights=None,xlabel="Reservoir OIP ($10^6$ bbls)",ylabel="",title="Average Reservoir OIP Realisations")
plt.ylim(0.0,135); plt.xlim([0,0.7])
plt.subplots_adjust(left=0.0, bottom=0.0, right=1.0, top=1.2, wspace=0.2, hspace=0.2)
plt.savefig("chart_iterations/montecarloReservoirOIP")
plt.show()
```
